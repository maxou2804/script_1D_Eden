```python
import plotly.graph_objects as go
import numpy as np

# Organize all data
# =============================================

p = [p_1, p_300, p_500, p_1000, p_5000, p_10000, p_20000]
beta = [beta_1, beta_300, beta_500, beta_1000, beta_5000, beta_10000, beta_20000]
alpha = [alpha_1, alpha_300, alpha_500, alpha_1000, alpha_5000, alpha_10000, alpha_20000]
names = ['1', '300', '500', '1000', '5000', '10000', '20000']

# =============================================
# Build interactive plot
# =============================================
fig = go.Figure()

for idx in range(7):
    x = np.array(p[idx])
    y = [idx] * len(x)
    z = beta[idx]

    fig.add_trace(go.Scatter3d(
        x=x, y=y, z=z,
        mode='lines+markers',
        name=f"Size {names[idx]}",
        marker=dict(size=5),
    ))

# =============================================
# Add transparent highlight box for   [0.28, 0.38]
# =============================================

# Find x-axis range (max probability across all datasets)
x_max = max([max(p[i]) for i in range(7)])

# Box corners
x_min, x_max = 0, x_max
y_min, y_max = 0, 6  # covers all 7 curves (0 to 6)
z_min, z_max = 0.28, 0.38

# 8 vertices of the cuboid
X = [x_min, x_max, x_max, x_min, x_min, x_max, x_max, x_min]
Y = [y_min, y_min, y_max, y_max, y_min, y_min, y_max, y_max]
Z = [z_min, z_min, z_min, z_min, z_max, z_max, z_max, z_max]

fig.add_trace(go.Mesh3d(
    x=X, y=Y, z=Z,
    opacity=0.25,
    color='yellow',
    flatshading=False,
    name=' range 0.280.38',
    showscale=False,
    alphahull=0
))

# =============================================
# Axis labels and formatting
# =============================================

fig.update_layout(
    scene=dict(
        xaxis_title='probabilities',
        yaxis_title='size',
        zaxis_title='beta value',
        yaxis=dict(
            tickmode='array',
            tickvals=list(range(7)),
            ticktext=names
        )
    ),
    width=1000,
    height=800,
    title='Interactive 3D Plot: Beta vs Probability vs Size (All Datasets)'
)

fig.show()
```
